With closer inspection, we might notice that 'rainfall', 'evaporation', 'windspeed9am', ' windspeed3pm',
and 'windgustspeed' may contain outliers

We will perform outliers removal later on

Now we can start by checking the missing values, aka nan value 

of each columns.

```python
print(weatherAUS.isnull().sum()) # missing value of each columns

# total missing value
total_missing_cnt = 0
for missing_cnt in weatherAUS.isnull().sum():
    total_missing_cnt = total_missing_cnt + missing_cnt
print()
print("Total missing count: {}".format(total_missing_cnt))
```

```python
# lets view the total percentage of each columns misssing value
for col, missing_cnt in zip(weatherAUS.columns,weatherAUS.isnull().sum()):
    print("{}, missing percentage: {}".format(col, round((missing_cnt/total_missing_cnt),4)))
```

As the inspection above, both categorical and numerical columns

contain missing values. The highest missing value percentage is belong to column 'sunshine'

```python
# lets handle the categorical variable missing value
categorical_col = weatherAUS.select_dtypes("object").columns

weatherAUS[categorical_col].isnull().sum()
```

```python
# for categorical missing value, we are just gonna drop it
weatherAUS.dropna(axis = 0, how = 'any', subset = categorical_col, inplace = True)

# check the missing values again
weatherAUS[categorical_col].isnull().sum()
```

```python
# lets now handle the numerical variable missing value
numerical_col = weatherAUS.select_dtypes(exclude = 'object').columns

# in this case, we will use median value to fill the nan value
for column in numerical_col:
    median = weatherAUS[column].median()
    weatherAUS[column].fillna(median, inplace = True)
```

For categorical column, we simply drop the row that contain missing value.

As for numerical column, we use median value as a replacement for each missing values.

Next, lets handle outliers in our dataset.

Since we have get some useful information from the 'describe' inspection above, we can

filter our the column that contains outliers.

```python
# settings for matplotlib
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 10}

plt.rc('font', **font)

# lets plot boxplot to visualize them
plt.figure(figsize = (10,7))

plt.subplot(2,3,1)
fig = sns.boxplot(x = weatherAUS['Rainfall'])
fig.set_ylabel('Rainfall')

plt.subplot(2,3,2)
fig = sns.boxplot(x = weatherAUS['Evaporation'])
fig.set_ylabel('Evaporation')

plt.subplot(2,3,3)
fig = sns.boxplot(x = weatherAUS['WindSpeed3pm'])
fig.set_ylabel('WindSpeed3am')

plt.subplot(2,3,4)
fig = sns.boxplot(x = weatherAUS['WindSpeed9am'])
fig.set_ylabel('WindSpeed9pm')

plt.subplot(2,3,5)
fig = sns.boxplot(x = weatherAUS['WindGustSpeed'])
fig.set_ylabel('WindGustSpeed')
plt.plot()
```

Next lets check the skewness of each of the columns. If there not well distributed, also

known as skew data, we can then use interquantile to perform the outliers removal.

```python
outlier_columns = ['Rainfall', 'Evaporation', 'WindGustSpeed', 'WindSpeed3pm', 'WindSpeed9am']

# settings for matplotlib
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 10}

plt.rc('font', **font)

plt.figure(figsize = (10,12))
for index, col in enumerate(outlier_columns):
    plt.subplot(2,3,index+1)
    plt.title("distribution of {}".format(col))
    sns.distplot(weatherAUS[col])
```

```python
weatherAUS.reset_index(inplace = True, drop = True)

# next we will use IQR techniques to remove the outliers
def outliers_removal(columns):
    drop_indexes = np.array([])
    for column in columns:
        lower_quantile = weatherAUS[column].quantile(0.25)
        upper_quantile = weatherAUS[column].quantile(0.75)
        drop_index = np.where(weatherAUS[column] > upper_quantile)[0]
        drop_indexes = np.append(drop_indexes,drop_index)
        print("The column {} has a lower quantile of {} and upper quantile of {}".format(column, lower_quantile, upper_quantile))
    drop_unique_index = np.unique(drop_indexes.astype(int))
    weatherAUS.drop(drop_unique_index, axis = 0, inplace = True)

outlier_columns = ['Rainfall', 'Evaporation', 'WindGustSpeed', 'WindSpeed3pm', 'WindSpeed9am']
outliers_removal(outlier_columns)
```

```python
# lets check if the outliers still present
for column in outlier_columns:
    print("Column {}: min value {}, max value {}".format(column, weatherAUS[column].min(), weatherAUS[column].max()))
```

Next, lets plot some graph to visualize our dataset.

```python
# lets visualize the mean of min temperate and max temperature of each location
avg_mintemp = weatherAUS.groupby("Location")["MinTemp"].mean()
avg_maxtemp = weatherAUS.groupby("Location")["MaxTemp"].mean()
```

```python
# since there is quite a number of location, we split into two figures, hence we make a simple function to plot
def min_max_temp(avg_mintemp, avg_maxtemp):
    fig = plt.figure(figsize = (25,8))
    ax = fig.add_subplot(111)
    ax2 = ax.twinx()

    width = 0.2
    avg_mintemp.plot(kind = "bar", ax = ax, color = 'blue', position = 0, width = width)
    avg_maxtemp.plot(kind = 'bar', ax = ax2, color = 'red', position = 1, width = width)
    plt.ylabel("Min and Max Temp")
    plt.xlabel("Locations")
    plt.plot()
```

```python
# settings for matplotlib
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 20}

plt.rc('font', **font)

half_len = len(avg_mintemp)//2
fh_mintemp, fh_maxtemp = avg_mintemp[:half_len], avg_maxtemp[:half_len]
sh_mintemp, sh_maxtemp = avg_mintemp[half_len:], avg_maxtemp[half_len:]

min_max_temp(fh_mintemp, fh_maxtemp)
min_max_temp(sh_mintemp, sh_maxtemp)
```

We need to perform feature scaling to scale the value of each

column into a reasonable range, 0 to 1. In this case, we will use standard scaler to do it.

```python
# define scaler
StandardScaler = preprocessing.StandardScaler
scaler = StandardScaler()

# scale the numerical columns
weatherAUS[numerical_col] = scaler.fit_transform(weatherAUS[numerical_col])
weatherAUS.head()
```

Next, lets perform one hot encoding on categorical columns.

```python
# in this case we will drop date column
weatherAUS.drop('Date', axis = 1, inplace = True)
categorical_col = categorical_col.drop(['Date','RainTomorrow'])

# raintomorrow is the y col so we pop it out
raintomorrow = weatherAUS.pop('RainTomorrow')
```

```python
# # define one hot encoder
onehotencoder = preprocessing.OneHotEncoder
ohe = onehotencoder(handle_unknown = 'error', sparse = False, drop = 'first')

categorical_column = weatherAUS[categorical_col]
weatherAUS.drop(columns = categorical_col, inplace = True)
ohe_cols = ohe.fit_transform(categorical_column)
```

```python
# then we concat the ohe cols back to our dataset
ohe_df = pd.DataFrame(ohe_cols, columns = ohe.get_feature_names())
weatherAUS.reset_index(inplace = True, drop = True)
weatherAUS = pd.concat([weatherAUS, ohe_df], axis = 1, sort = False)
```

```python
weatherAUS.head()
```

Lastly, we can now split into train and test dataset and build the

logistic regression model and train it with the cleaned data.

```python
# lets see y data first
print(raintomorrow)
```

```python
raintomorrow.reset_index(inplace = True, drop = True) # reset the index first

# convert the string value into 0 or 1
# if no rain tomorrow - 0 else 1
y = raintomorrow.apply(lambda x: 0 if x == 'No' else 1)
x = weatherAUS
```

```python
# split train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 10)
```

```python
# define logistic regression model
lg_model = LogisticRegression(penalty = 'l2', C = 2.0, tol = 1e-4)
lg_model.fit(x_train, y_train)
```

Lets evaluate the performance of the logistic regression model.

```python
y_pred = lg_model.predict(x_test)
```

```python
accuracy_score = metrics.accuracy_score(y_pred,y_test)
print(accuracy_score)
```

```python
def print_confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=14):
    df_cm = pd.DataFrame(
        confusion_matrix, index=class_names, columns=class_names, 
    )
    fig = plt.figure(figsize=figsize)
    
    heatmap = sns.heatmap(df_cm, annot=True, fmt="d")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('Truth')
    plt.xlabel('Prediction')
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
print_confusion_matrix(confusion_matrix, ['NoRain','Rain'])
```
